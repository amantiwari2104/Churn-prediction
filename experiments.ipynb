```python
import pandas as pd
import tensorflow as tf
import pickle
```

```python
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
data = pd.read_csv('C:\\Users\\aman tiwari\\OneDrive\\Desktop\\Ann classification\\Churn_Modelling.csv')
```

```python
data.head()
```

```python
#preprocess the data
data = data.drop(['RowNumber','CustomerId','Surname'],axis=1)
```

```python
data.head()
```

```python
#Encode categorical variable
laber_encoder_gender=LabelEncoder()
data['Gender'] = laber_encoder_gender.fit_transform(data['Gender'])
```

```python
data.head()
```

```python
#onehotencode geography column
from sklearn.preprocessing import OneHotEncoder
onehot_encoder_geo = OneHotEncoder()
from sklearn.compose import ColumnTransformer
preprocessor = ColumnTransformer(
    [
        ("OneHotEncoder", onehot_encoder_geo, ['Geography']),
    ], remainder='passthrough'             #passthrough is used to reatin all the other columns
)
```

```python
data_processed=preprocessor.fit_transform(data)     #this transforms the data into a numpy array
```

```python
data
```

```python
ohe_feature_names = preprocessor.named_transformers_['OneHotEncoder'].get_feature_names_out(['Geography'])
#.get_feature_names_out(['Geography']) returns the actual names of the columns created by OneHotEncoding.
final_columns = list(ohe_feature_names) + list(data.columns.drop('Geography'))
#combines the list of one hot encoded columns and the rest of columns from previous data except geography

processed_data= pd.DataFrame(data_processed, columns=final_columns)
#creates a data frame of new data
print(processed_data.head())
```

```python
#save the encoders 
with open('laber_encoder_gender.pkl','wb') as file:
    pickle.dump(laber_encoder_gender,file)

with open('preprocessor.pkl','wb') as file:
    pickle.dump(preprocessor,file)   
```

```python
processed_data.head()
```

```python
x = processed_data.drop('Exited',axis=1)
y = processed_data['Exited']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state=42)

#scale these 
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
with open('scaler.pkl','wb') as file:
    pickle.dump(scaler,file)
```

#### ANN Implementation 

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
import datetime 
```

```python
#Build our ANN model
model = Sequential([
    Dense(64,activation="relu",input_shape=(x_train.shape[1],)), #HL1
    #x_train.shape[1] → number of features (columns)
    Dense(32,activation="relu"), #HL2
    Dense(1,activation="sigmoid")  #output layer 
]
)
```

```python
#compile the model 
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
##set up the tensorboard
log_dir = "logs/fit" +datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorflow_callback = TensorBoard(log_dir=log_dir,histogram_freq=1)
```

```python
#Setup Early stopping
#If are certain epochs our model is trained at good accuracy we can stop the training
early_stopping_callback = EarlyStopping(monitor='val_loss',patience=5,restore_best_weights=True)

#wait till atleast 5epochs if there are no changes and if there are no changes get the best weights
```

```python
#train the model
history = model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=100,
                    callbacks=[tensorflow_callback,early_stopping_callback])
```

```python
print(model.summary())
model.save('model.keras')


#saving the model and done in h5 file because it is compatible with keras
```
